# Loading a dataset

Seaborn comes with several built-in datasets for demonstration purposes. You can load one of these datasets using the load_dataset() function:

```python
sns.get_dataset_names()
df = sns.load_dataset("iris")
df.head()
df
```

# Scatter Plot

- When to use: Scatter plots are used when you want to understand the relationship between two continuous variables, usually to find trends or correlations.
- How it works: A scatter plot displays data points on a Cartesian plane, with each axis representing one of the two variables. Each point represents a single observation.
- Why to choose it: Choose a scatter plot when you want to visualize the relationship between two continuous variables or identify potential outliers.
- How to read: Look for clusters, trends, or patterns between the two variables. If there's a positive or negative trend, it might indicate a correlation between the variables.

```python
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
```

## What about petals?

```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
```

# Histogram

- When to use: Histograms are used to visualize the distribution of a single continuous variable, allowing you to see the shape, central tendency, and dispersion of the data.
- How it works: A histogram divides the data into intervals (bins) and plots the frequency of observations in each bin as a bar.
- Why to choose it: Choose a histogram when you want to visualize the distribution of a single continuous variable or identify potential outliers or skewness in the data.
- How to read: Look at the shape of the bars to understand the distribution of the data, whether it's symmetric, skewed, or multimodal.

```python
sns.histplot(data=df, x="sepal_length", kde=True)
```

```python
sns.histplot(data=df, x="petal_length", kde=True)
```

# Box Plot
- When to use: Box plots are used to visualize the distribution and spread of a single continuous variable, as well as to identify potential outliers.
- How it works: A box plot displays a box with the lower quartile (Q1), median (Q2), and upper quartile (Q3) as its borders. Whiskers extend from the box to the minimum and maximum values within 1.5 times the interquartile range (IQR). Outliers are plotted as individual points.
- Why to choose it: Choose a box plot when you want to visualize the distribution, spread, and potential outliers of a single continuous variable or to compare the distributions of multiple variables.
- How to read: Look at the box and whiskers to assess the distribution and spread of the data. Outliers are plotted as individual points outside the whiskers.

```python
sns.boxplot(data=df, x="species", y="sepal_length")
```

```python
sns.boxplot(data=df, x="species", y="petal_length")
```

# Pair Plot
- When to use: Pair plots are used to visualize the relationships between multiple continuous variables in a single plot, often to explore correlations or trends.
- How it works: A pair plot is a grid of scatter plots that display each pair of variables against each other. The diagonal often shows a histogram or kernel density plot of each variable.
- Why to choose it: Choose a pair plot when you want to explore relationships between multiple continuous variables or to identify potential correlations or trends.
- How to read: Examine each scatter plot to find trends, correlations, or patterns between pairs of variables. Histograms or kernel density plots along the diagonal show the distribution of each variable.

```python
sns.pairplot(data=df, hue="species")
plt.show()
```

# Heat Map
- When to use: Heat maps are used to visualize the relationship between two categorical variables, or to display data on a grid where colors represent values.
- How it works: A heat map represents data in a matrix, where each cell has a color that corresponds to the value of the data point. The color scale can be continuous or discrete, with a legend indicating the range of values represented by each color.
- Why to choose it: Choose a heat map when you want to visualize the relationship between two categorical variables, or to display data on a grid where colors represent values. Heat maps are especially useful when dealing with large datasets or when you want to identify patterns and trends easily.
- How to read: Look at the color scale to understand the range of values represented by each color. Examine the matrix to find patterns, trends, or clusters in the data. High or low values will be represented by distinct colors, making it easy to spot areas of interest or concern.

```python
correlation_matrix = df.corr()
# correlation_matrix 
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

# Customizing Plots

```python
# Customizing plots
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", style="species", size="petal_length")
plt.title("Iris Dataset Scatterplot")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

# Line Plot using the FMRI Dataset
- When to use: Line plots are typically used to visualize time series data or to show how a continuous variable changes over time.
- How it works: A line plot displays data points connected by lines on a Cartesian plane. The x-axis usually represents time, while the y-axis represents the variable of interest.
- Why to choose it: Choose a line plot when you want to visualize the progression of a continuous variable over time or to identify trends and patterns.
- How to read: Look for upward or downward trends, periodic patterns, or sudden changes in the plotted line.

```python
data = sns.load_dataset("fmri")
data
```

```python
# Create a lineplot
sns.lineplot(data=data, x="timepoint", y="signal", hue="region")

# Customize the plot
plt.title("fMRI Signal Change Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal Change")

# Plot it
plt.show()
```
