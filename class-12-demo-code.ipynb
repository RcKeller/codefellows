# Pandas DataFrame

A DataFrame is a two-dimensional labeled data structure with columns that can be of different types (similar to a table in Excel). You can think of a DataFrame as a collection of Series objects. To create a DataFrame, use the pd.DataFrame() function:

```python
data = {'Name': ['Alice', 'Bob', 'Carol'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

df
```

# Ask Questions and Find Answers!

What's best low sugar option cereal? 
aka What's the highest rated cereal with less than 5 grams of sugar?


```python
# Create subset of dataframe following an expression
low_sugar = df[ (df.sugars < 5) & (df.sugars > 0) ]


type(low_sugar) # type is DataFrame
low_sugar.head() # can still do DataFrame methods on the sub-dataframe

# Let's sort!
# low_sugar.sort_values(by="rating", ascending=False).head(1).name
low_sugar = low_sugar.sort_values(by="rating", ascending=False)
low_sugar.describe()
```

# Which manufacturer has the highest average sugar?

```python
# group by manufacturer and calculate mean sugar for each group
sugar_by_mfr = df.groupby('mfr')['sugars'].mean()

sugar_by_mfr.head()

sugar_by_mfr.sort_values(ascending=False).head(1)
```
