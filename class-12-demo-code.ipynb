# Pandas SeriesÂ¶

A Pandas Series is a one-dimensional labeled array. Think of it as a single column in an Excel spreadsheet or a single list with labels. To create a Series, use the pd.Series() function:

```python
data = pd.Series([3, 5, 7, 9])

data
```

# Pandas DataFrame

A DataFrame is a two-dimensional labeled data structure with columns that can be of different types (similar to a table in Excel). You can think of a DataFrame as a collection of Series objects. To create a DataFrame, use the pd.DataFrame() function:

```python
data = {'Name': ['Alice', 'Bob', 'Carol'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

df
```

# Selecting and Slicing Data
Pandas provides various ways to select and slice data from a DataFrame. Here are some examples:

- Select a single column: df['Name']
- Select multiple columns: df[['Name', 'Age']]
- Select rows by index: df.loc[0] or df.iloc[0]
- Slice rows: df[1:3]

```python
df.iloc[54]
```
